## Introdução

Este notebook apresenta os resultados de uma pesquisa realizada com 20 pessoas sobre o tema do uso do **vape (cigarro eletrônico)**. Os dados foram coletados por meio de perguntas simples e visam entender o nível de contato e curiosidade dos participantes em relação ao uso do vape.

---

## Gráfico 1 - Você já fez uso do cigarro eletrônico?

#### 1.1 - Amostragem

- **Sim:** 40%
- **Não:** 60%

#### 1.2 - Gráfico


```python
import matplotlib.pyplot as plt

labels_fumo = ['Sim', 'Não']
sizes_fumo = [40, 60]
colors = ['#3C95FA','#FA5C3C']

plt.figure(figsize=(6,6))
plt.pie(sizes_fumo, labels=labels_fumo, autopct='%1.1f%%', colors=colors, wedgeprops=dict(width=0.5))
plt.axis('equal') 
plt.title('Gráfico 1 - Você já fez uso do cigarro eletrônico?')

plt.savefig("graphs/grafico1.png", format="png")

plt.show()
```

#### 1.3 - Análise

Neste gráfico observamos que 60% dos participantes nunca fumaram, enquanto 40% afirmaram já ter fumado alguma vez na vida. Isso demonstra que, apesar do acesso ao cigarro ainda ser presente, a maioria da amostra parece evitar o contato com o tabaco/vape.

---

## Gráfico 2 - Por que você usaria cigarro eletrônico?

#### 2.1 - Amostragem 

- **Curiosidade:** 25%
- **Influência:** 35%
- **Estresse:** 20%
- **Outros:** 20%

#### 2.2 - Gráfico

```python
import matplotlib.pyplot as plt

labels_usage = ['Curiosidade','Influência', 'Estresse', 'Outros']
sizes_usage = [25, 35, 20, 20]
colors = ['#FF6547', '#E6FF47', '#479AFF', '#6A7282']

plt.figure(figsize=(6,6))
plt.pie(sizes_usage, labels=labels_usage, autopct='%1.1f%%', colors=colors, wedgeprops=dict(width=0.5))
plt.title('Gráfico 2 - Por que você usaria cigarro eletrônico?')

plt.savefig("graphs/grafico2.png", format="png")

plt.show()
```

#### 2.3 - Análise

O gráfico revela que influência social é o principal fator no interesse pelo cigarro eletrônico, seguido de curiosidade. Essa informação pode ser fundamental para campanhas educativas, mostrando que muitas decisões relacionadas ao uso do vape não são baseadas em necessidade pessoal, mas sim em fatores externos.
